## Seaborn

```python
#dist plot
#joint plot : used for univariate data use it when you have one cont var
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
tips['sex'].value_counts()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30);
#kernel density distribution 
```

```python
#joint plot : used for bivariate data use it when you have two cont var
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde');
```

```python
#pairplot
#It plots pairwise relationship wise ENTIRE DATAFRAME
# It also supports color hue argumenet for categoriacal data
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex');
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm');
```

## Seaborn Categorical Plots

```python
#barplot: It is a general plot that help us to aggregate the data
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
#countplot: same as barplot except estimator is explicitly counting number of occurence only x is passed
sns.countplot(x='sex',data=tips);
```

```python
#barplot/box and whisker plot
#it shows distribution of quantitaive data and helps in identifying the outliers
#x is categorical point
#y is a numeric point

sns.boxplot(data = tips,palette='rainbow',orient='h')
```
